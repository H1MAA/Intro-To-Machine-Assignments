```python
def leastSquares(x, y):
    n = len(x)
    
    sum_x = sum(x)
    sum_y = sum(y)
    sum_x2 = sum(x_i**2 for x_i in x)
    sum_xy = sum(x_i*y_i for x_i, y_i in zip(x, y))

    m = (n*sum_xy - sum_x*sum_y) / (n*sum_x2 - sum_x**2)
    b = (sum_y - m*sum_x) / n
    return m, b
```

```python
x = [-1, 0 , 1, 2]
y = [0, 2, 4, 5]

m, b = leastSquares(x, y)
print(f"y = {m}x + {b}")

import matplotlib.pyplot as plt
import numpy as np

x = np.array(x)
y = np.array(y)
plt.scatter(x, y)
plt.plot(x, m*x + b)
plt.show()
```
